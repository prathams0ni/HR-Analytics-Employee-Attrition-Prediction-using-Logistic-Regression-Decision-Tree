### Model Building on the basis of Scaling:

### Spliting Data:

```python
features = ['Satisfaction Level', 'Last Evaluation', 'Number of Projects',
       'Monthly Hours', 'Total Time at the Company', 'Work Accidents',
       'Promoted in Last 5 yrs', 'Departments', 'Salary', 'Management']
```

```python
X = hr_df[features]
Y = hr_df['Quit the Company']
print(f"X shape: {X.shape}")
print(f"Y shape: {Y.shape}")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

### Feature Scalling:

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
print("Feature scaling completed")
print("Scaled training data shape:", X_train_scaled.shape)
print("Scaled test data shape:", X_test_scaled.shape)
```

```python
# Show mean and std after scaling (should be ~0 and ~1)
print("Training data after scaling (mean ~0, std ~1):")
print("Mean:", np.mean(X_train_scaled, axis=0).round(2))
print("Std:", np.std(X_train_scaled, axis=0).round(2))
```

### Build and Train the Logistic Regression Model:

```python
logreg = LogisticRegression()
logreg.fit(X_train_scaled, Y_train)
```

### Making Prediction:

```python
Y_pred = logreg.predict(X_test_scaled)
print(Y_pred)
Y_pred_proba = logreg.predict_proba(X_test_scaled)
print(Y_pred_proba)
```

```python
print(f"First 10 predictions: {Y_pred[:10]}")
print(f"First 10 actual values: {Y_test.values[:10]}")
```

### Model Performance:

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")
print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(Y_test,Y_pred)
sns.heatmap(cfm, annot=True, fmt='g', cbar=False, cmap='BuPu')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()

print("Classification report:")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test,Y_pred)
print("Accuracy of the model: ",acc)
```

### Find the score of training & testing data:

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(logreg.score(X_train, Y_train)))

print('Test set score: {:.4f}'.format(logreg.score(X_test, Y_test)))
```

### Model Tuning:

```python
y_pred_prob = logreg.predict_proba(X_test_scaled)
y_pred_prob[:,1]
```

```python
for a in np.arange(0.4,0.61,0.01):
    predict_mine = np.where(y_pred_prob[:,1] > a, 1, 0)
    cfm=confusion_matrix(Y_test, predict_mine)
    total_err=cfm[0,1]+cfm[1,0]
    print("Errors at threshold ", a, ":",total_err, " , type 2 error :",
          cfm[1,0]," , type 1 error:", cfm[0,1])
```

```python
y_pred_class=[]
for value in y_pred_prob[:,1]:
    if value > 0.4:
        y_pred_class.append(1)
    else:
        y_pred_class.append(0)
print(y_pred_class)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(Y_test,y_pred_class)
sns.heatmap(cfm, annot=True, fmt='g', cbar=False, cmap='BuPu')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()


print("Classification report:")

print(classification_report(Y_test,y_pred_class))

acc=accuracy_score(Y_test,y_pred_class)
print("Accuracy of the model: ",acc)
```

### Creating a Testing Unseen Data:

```python
import pandas as pd
import numpy as np
import base64
from IPython.display import HTML

def create_download_link(df, filename="HR_test_15000_samples.csv", link_text="Download CSV File"):
    """Generate a download link for the DataFrame"""
    csv = df.to_csv(index=False)
    b64 = base64.b64encode(csv.encode()).decode()  # Encode to base64
    href = f'<a href="data:file/csv;base64,{b64}" download="{filename}" target="_blank">{link_text}</a>'
    return href

# Set random seed for reproducibility
np.random.seed(42)

# Define department distribution (based on original data patterns)
departments = ['sales', 'technical', 'support', 'IT', 'RandD', 'product_mng', 'marketing', 'accounting', 'hr', 'management']
dept_weights = [0.25, 0.15, 0.12, 0.10, 0.08, 0.08, 0.07, 0.07, 0.05, 0.03]

# Define salary distribution
salaries = ['low', 'medium', 'high']
salary_weights = [0.50, 0.40, 0.10]

print("Generating 15,000 test samples...")

# Generate 15,000 samples
n_samples = 15000
data = []

for i in range(n_samples):
    if i % 3000 == 0:  # Progress indicator
        print(f"Progress: {i}/{n_samples} samples generated...")
    
    # Generate correlated satisfaction and evaluation
    satisfaction = np.random.beta(2, 5)  # Skewed left like original data
    
    # Evaluation correlated with satisfaction but with some noise
    evaluation = min(1.0, max(0.3, satisfaction + np.random.normal(0, 0.15)))    
    evaluation = round(evaluation, 2)
    satisfaction = round(satisfaction, 2)
    
    # Projects (2-7, with some correlation to evaluation)
    if evaluation > 0.8:
        projects = np.random.choice([4, 5, 6, 7], p=[0.2, 0.4, 0.3, 0.1])
    elif evaluation > 0.6:
        projects = np.random.choice([3, 4, 5, 6], p=[0.3, 0.4, 0.2, 0.1])
    else:
        projects = np.random.choice([2, 3, 4], p=[0.6, 0.3, 0.1])
    
    # Monthly hours (correlated with projects)
    base_hours = 140
    hours = base_hours + (projects * 25) + np.random.normal(0, 20)
    hours = max(120, min(350, int(hours)))
    
    # Time at company (1-6 years)
    time_at_company = np.random.choice([1, 2, 3, 4, 5, 6], p=[0.05, 0.15, 0.30, 0.25, 0.15, 0.10])
    
    # Work accidents (rare)
    work_accidents = 1 if np.random.random() < 0.02 else 0
    
    # Quit company (mostly 1 for this test set, but some 0)
    quit_company = 1 if np.random.random() < 0.85 else 0
    
    # Promotions (rare)
    promoted = 1 if np.random.random() < 0.02 else 0
    
    # Department
    department = np.random.choice(departments, p=dept_weights)
    
    # Salary (correlated with department and evaluation)
    if department == 'management' or evaluation > 0.85:
        salary = np.random.choice(salaries, p=[0.2, 0.5, 0.3])
    elif evaluation > 0.7:
        salary = np.random.choice(salaries, p=[0.4, 0.5, 0.1])
    else:
        salary = np.random.choice(salaries, p=[0.7, 0.3, 0.0])
    
    # Management ID (1-100, with some departments having specific ranges)
    if department == 'management':
        management_id = np.random.randint(1, 20)
    elif department in ['sales', 'technical']:
        management_id = np.random.randint(20, 60)
    else:
        management_id = np.random.randint(60, 101)
    
    data.append([
        satisfaction, evaluation, projects, hours, time_at_company,
        work_accidents, quit_company, promoted, department, salary, management_id
    ])

print("Dataset generation completed!")
print("Creating DataFrame...")

# Create DataFrame
columns = [
    'Satisfaction Level', 'Last Evaluation', 'Number of Projects', 'Monthly Hours',
    'Total Time at the Company', 'Work Accidents', 'Quit the Company',
    'Promoted in Last 5 yrs', 'Departments', 'salary', 'Management'
]

df = pd.DataFrame(data, columns=columns)

# Display summary statistics
print("\n" + "="*50)
print("DATASET SUMMARY")
print("="*50)
print(f"Total samples: {len(df):,}")
print(f"Departments distribution:")
print(df['Departments'].value_counts())
print(f"\nSalary distribution:")
print(df['salary'].value_counts())
print(f"\nKey Statistics:")
print(f"Average satisfaction: {df['Satisfaction Level'].mean():.3f}")
print(f"Average evaluation: {df['Last Evaluation'].mean():.3f}")
print(f"Average projects: {df['Number of Projects'].mean():.1f}")
print(f"Average monthly hours: {df['Monthly Hours'].mean():.1f}")
print(f"Employees who quit: {df['Quit the Company'].sum():,}")
print(f"Promotions: {df['Promoted in Last 5 yrs'].sum():,}")
print(f"Work accidents: {df['Work Accidents'].sum():,}")

print("\n" + "="*50)
print("DOWNLOAD LINK")
print("="*50)

# Generate and display download link
download_link = create_download_link(
    df, 
    filename="HR_test_15000_samples.csv", 
    link_text="ðŸ“¥ Click here to download the 15,000 sample test dataset"
)

# Display the download link
display(HTML(download_link))

print("\nFirst 5 rows of the dataset:")
print(df.head())

# Also save locally as backup
df.to_csv('HR_test_15000_samples.csv', index=False)
print(f"\nâœ… Dataset also saved locally as 'HR_test_15000_samples.csv'")
```
